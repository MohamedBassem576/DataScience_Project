```python

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Convert categorical variables to numeric using LabelEncoder
categorical_cols = ['work_year', 'experience_level', 'employment_type', 'job_title',
                    'salary_currency', 'employee_residence', 'company_location', 'company_size', 'salary_in_usd']
label_encoder = LabelEncoder()
for col in categorical_cols:
    df[col] = label_encoder.fit_transform(df[col])
df2=df
df2=df2.drop("salary",axis=1)
df2=df2.drop("salary_currency",axis=1)

# Split the data into features and output
features = df2.drop("company_size", axis=1)
output = df2["company_size"]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, output, test_size=0.3, random_state=1)

df.dtypes
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Apply the K-Nearest Neighbors algorithm
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
knn_pred = knn.predict(X_test) 
print(knn_pred)
print(y_test.to_xarray())
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score

# Compute the evaluation metrics for KNN
knn_acc = accuracy_score(y_test, knn_pred)
knn_cm = confusion_matrix(y_test, knn_pred)
knn_prec = precision_score(y_test, knn_pred, average='weighted', zero_division=1)
knn_rec = recall_score(y_test, knn_pred, average='weighted', zero_division=1)

# Print the evaluation metrics for KNN
print("KNN Accuracy:", knn_acc)
print("KNN Precision:", knn_prec)
print("KNN Recall:", knn_rec)

# confusion matrix.
sns.heatmap(knn_cm,
            annot=True,
            fmt='g',
            xticklabels=['Small','Medium', 'Large'],
            yticklabels=['Small','Medium',"Large"])
plt.ylabel('Prediction',fontsize=13)
plt.xlabel('Actual',fontsize=13)
plt.title('Confusion Matrix',fontsize=17)
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score

# Create an instance of the Naive Bayes classifier
nb = GaussianNB()

# Train the model
nb.fit(X_train, y_train)

# Make predictions on the test set
nb_predictions = nb.predict(X_test)

# Compute the evaluation metrics for Naive Bayes
nb_accuracy = accuracy_score(y_test, nb_predictions)
nb_confusion_matrix = confusion_matrix(y_test, nb_predictions)
nb_precision = precision_score(y_test, nb_predictions, average='weighted', zero_division=1)
nb_recall = recall_score(y_test, nb_predictions, average='weighted', zero_division=1) 

# Print the evaluation metrics for Naive Bayes
print("Naive Bayes Accuracy:", nb_accuracy)
print("Naive Bayes Precision:", nb_precision)
print("Naive Bayes Recall:", nb_recall)
# confusion matrix.
sns.heatmap(nb_confusion_matrix,
            annot=True,
            fmt='g',
            xticklabels=['Small','Medium', 'Large'],
            yticklabels=['Small','Medium',"Large"])
plt.ylabel('Prediction',fontsize=13)
plt.xlabel('Actual',fontsize=13)
plt.title('Confusion Matrix',fontsize=17)
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score

# Decision Trees
decision_tree = DecisionTreeClassifier()

# Train the models
decision_tree.fit(X_train, y_train)

# Make predictions on the test set
decision_tree_predictions = decision_tree.predict(X_test)

# Evaluate the performance of the models
decision_tree_accuracy = accuracy_score(y_test, decision_tree_predictions)
decision_tree_confusion_matrix = confusion_matrix(y_test, decision_tree_predictions)
decision_tree_precision = precision_score(y_test, decision_tree_predictions, average='weighted', zero_division=0)
decision_tree_recall = recall_score(y_test, decision_tree_predictions, average='weighted', zero_division=0)

# Print the evaluation metrics for Decision Tree
print("Decision Tree Accuracy:", decision_tree_accuracy)
print("Decision Tree Precision:", decision_tree_precision)
print("Decision Tree Recall:", decision_tree_recall)
# confusion matrix.
sns.heatmap(decision_tree_confusion_matrix,
            annot=True,
            fmt='g',
            xticklabels=['Small','Medium', 'Large'],
            yticklabels=['Small','Medium',"Large"])
plt.ylabel('Prediction',fontsize=13)
plt.xlabel('Actual',fontsize=13)
plt.title('Confusion Matrix',fontsize=17)
plt.show()
```
